```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram
import matplotlib
```

# Understanding Quantum Circuits
In a circuit, we typically need to do three jobs: First, encode the input, then do some actual computation, and finally extract an output.

```python
# a circuit with eight qubits and eight outputs.
qc_output = QuantumCircuit(8)
```

```python
# extraction of outputs
qc_output.measure_all()
```

```python
qc_output.draw(initial_state=True) 
```

Since we are not doing anything quantum in the above circuit, the result will always be same, i.e 00000000

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_output).result()
counts = result.get_counts()
plot_histogram(counts)
```

The probability of getting 00000000 is 1

# NOT Gate in Qiskit

```python
qc_encode = QuantumCircuit(8)
# x is used for NOT gate
# Also, the bits in qiskit start from right to left
qc_encode.x(7)
qc_encode.draw(initial_state=True)
```

```python
#Extracting results and drawing
qc_encode.measure_all()
qc_encode.draw(initial_state=True)
```

Please note that the diagram has 3 parts, 
* the first is the initial state (represented by |0> in the beginning)
* second part represents operations or gates (x is represented by a 'X' in square)
* third is the output which is showing [M] since it's not determined yet.

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_encode).result()
counts = result.get_counts()
plot_histogram(counts)
```

Since we are talking about quantums, we can only talk in probablities, here probability of the output to be 10000000 is 1.

# C-NOT Gate

Short for controlled not gate. It operates on two qbits, the first is known as controlled qbit and second is output qbit. It flips the target qbit if the controlled qbit is 1.

### Matrix:

controlled qbit | target qbit  

0 | 0 -> 0 | 0  
0 | 1 -> 0 | 1  
1 | 0 -> 1 | 1  
1 | 1 -> 1 | 0  

```python
qc_cnot = QuantumCircuit(2)
# first argument is controlled qbit and second is target
qc_cnot.cx(0,1)
qc_cnot.draw()
```

```python
qc_cnot.measure_all()
qc_cnot.draw(initial_state=True)
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_cnot).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
qc = QuantumCircuit(2)
qc.x(0)
qc.cx(0,1)
qc.measure_all()
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc).result()
counts = result.get_counts()
plot_histogram(counts)
```

# Toffoli Gate (AND Gate)

This works on two controlled qbits and one target bit. This will perform a NOT on the target qubit only when both controls are in state 1.

```python
qc_ha = QuantumCircuit(3)
# use ccx to write the AND of the inputs on qubit 3
qc_ha.ccx(0,1,2)
qc_ha.draw()
```

```python
qc_ha.measure_all()
qc_ha.draw()
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_ha).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
qc_ha2 = QuantumCircuit(3)
qc_ha2.x(1)
qc_ha2.x(0)
# use ccx to write the AND of the inputs on qubit 3
qc_ha2.ccx(0,1,2)
qc_ha2.draw()
```

```python
qc_ha2.measure_all()
qc_ha2.draw()
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_ha2).result()
counts = result.get_counts()
plot_histogram(counts)
```
